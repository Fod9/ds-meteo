### Question 6

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(df['Temperature'], bins=20)
plt.title('Histogram des températures')
plt.xlabel('Temperature')
plt.ylabel('Frequence')
plt.show()
```
